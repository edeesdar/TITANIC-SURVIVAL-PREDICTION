```python
def checkbyname(name):
    if name in df["Name"].values:
       
        index = df.index[df["Name"] == name].tolist()
        if df.loc[index, "Survived"].values[0] == 1:
            print(f"The name '{name}' survived.")
        else:
            print(f"The name '{name}'  did not survive.")
    else:
        print(f"The name '{name}' is not present in the dataset.")
```
